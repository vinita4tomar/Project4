```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Load the data from CSV file
data = pd.read_csv('parsed_spotify_data.csv')
# Display the first few rows of the dataset
print("First few rows of the dataset:")
data.head()
```

```python
# Preprocess the data
X = data.select_dtypes(include=['number'])  # Select only numerical columns

# Choose the number of clusters (you may need to tune this parameter)
num_clusters = 3

# Create a KMeans instance
kmeans = KMeans(n_clusters=num_clusters, random_state=42)

# Fit the KMeans model to the data
kmeans.fit(X)

# Get the cluster labels
cluster_labels = kmeans.labels_
cluster_labels
```

```python
# Add the cluster labels to the original dataset
data['cluster'] = cluster_labels

# Display the count of records in each cluster
print("\nCount of records in each cluster:")
print(data['cluster'].value_counts())

```

```python
# Visualize the clusters (using the first two numerical features)
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=cluster_labels, cmap='viridis')
plt.xlabel(X.columns[0])
plt.ylabel(X.columns[1])
plt.title('Clustering Visualization')
plt.show()
```

```python
import pandas as pd
from itertools import combinations

# Load the data from CSV file
data = pd.read_csv('parsed_spotify_data.csv')

# Get the list of column names
columns = data.columns

# Generate all possible combinations of features for clustering
all_combinations = []
for r in range(1, len(columns)+1):
    combinations_r = combinations(columns, r)
    all_combinations.extend(combinations_r)

# Display all possible combinations
for i, combination in enumerate(all_combinations, 1):
    print(f"Combination {i}: {combination}")
```
