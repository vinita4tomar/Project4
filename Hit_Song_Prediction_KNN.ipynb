```python
import pandas as pd
from pathlib import Path
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
spotify_2023 = pd.read_csv("Resources/spotify_2023.csv", encoding='ISO-8859-1')

spotify_2023.head(10)
```

```python
spotify_2023 = spotify_2023.dropna()
```

```python
for index, row in spotify_2023.iterrows():
    if row['streams'] >= 100000000:
        spotify_2023.at[index, 'hit_song'] = "Hit"
    else:
        spotify_2023.at[index, 'hit_song'] = "Not Hit"
```

```python
spotify_2023.head(10)
```

```python
spotify_2023.tail(10)
```

```python
y =spotify_2023['hit_song']
X =spotify_2023.drop(columns='hit_song')
```

```python
X.head()
```

```python
y[:5]
```

```python
X = pd.get_dummies(X)
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
scaler = StandardScaler()

X_scaler = scaler.fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
model = KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)


y_pred
```

```python
confusion_matrix(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```
